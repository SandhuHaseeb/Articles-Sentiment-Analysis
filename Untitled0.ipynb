```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Articles Sentiment Analysis/Sentiment_dataset - short.csv')

# Ensure the text column and sentiment column exist
if not {'text', 'sentiment'}.issubset(df.columns):
    raise ValueError("Dataset must contain 'text' and 'sentiment' columns")

# Preprocess the data
df['text'] = df['text'].astype(str)

# Inspect the 'sentiment' column to find valid numeric values
print("Unique values in the 'sentiment' column before filtering:")
print(df['sentiment'].unique())

# Attempt to convert 'sentiment' column to numeric, coerce errors to NaN
df['sentiment'] = pd.to_numeric(df['sentiment'], errors='coerce')

# Remove rows with NaN values in the 'sentiment' column
df.dropna(subset=['sentiment'], inplace=True)

# Convert sentiment to integer
df['sentiment'] = df['sentiment'].astype(int)

# Ensure only 0 and 1 are present in the 'sentiment' column
df = df[df['sentiment'].isin([0, 1])]

# Print the number of valid rows
print(f"Number of valid rows: {len(df)}")

# Check for NaN values and remove them
df.dropna(subset=['text'], inplace=True)

# Print the number of rows after removing NaNs
print(f"Number of rows after removing NaNs: {len(df)}")

# Extract the text and sentiment values
X = df['text'].values
y = df['sentiment'].values

# Check if there are any samples left
if len(X) == 0 or len(y) == 0:
    raise ValueError("No valid samples found in the dataset.")

# Tokenize the text
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(X)
X_tokenized = tokenizer.texts_to_sequences(X)

# Pad the sequences
maxlen = 200
X_padded = pad_sequences(X_tokenized, maxlen=maxlen)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)

# Print the shapes of the training and testing sets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

# Build the CNN model
model = Sequential([
    Embedding(input_dim=10000, output_dim=128, input_length=maxlen),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    MaxPooling1D(pool_size=5),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train the model
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.2f}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D, Dense, Dropout, LSTM
from tensorflow.keras.callbacks import EarlyStopping

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Articles Sentiment Analysis/Sentiment_dataset - short.csv')

# Ensure the text column and sentiment column exist
if not {'text', 'sentiment'}.issubset(df.columns):
    raise ValueError("Dataset must contain 'text' and 'sentiment' columns")

# Preprocess the data
df['text'] = df['text'].astype(str)

# Inspect the 'sentiment' column to find valid numeric values
print("Unique values in the 'sentiment' column before filtering:")
print(df['sentiment'].unique())

# Attempt to convert 'sentiment' column to numeric, coerce errors to NaN
df['sentiment'] = pd.to_numeric(df['sentiment'], errors='coerce')

# Remove rows with NaN values in the 'sentiment' column
df.dropna(subset=['sentiment'], inplace=True)

# Convert sentiment to integer
df['sentiment'] = df['sentiment'].astype(int)

# Ensure only 0 and 1 are present in the 'sentiment' column
df = df[df['sentiment'].isin([0, 1])]

# Print the number of valid rows
print(f"Number of valid rows: {len(df)}")

# Check for NaN values and remove them
df.dropna(subset=['text'], inplace=True)

# Print the number of rows after removing NaNs
print(f"Number of rows after removing NaNs: {len(df)}")

# Extract the text and sentiment values
X = df['text'].values
y = df['sentiment'].values

# Check if there are any samples left
if len(X) == 0 or len(y) == 0:
    raise ValueError("No valid samples found in the dataset.")

# Tokenize the text
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(X)
X_tokenized = tokenizer.texts_to_sequences(X)

# Pad the sequences
maxlen = 200
X_padded = pad_sequences(X_tokenized, maxlen=maxlen)

# Split the data into training, validation, and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)  # 0.25 x 0.8 = 0.2

# Print the shapes of the training, validation, and testing sets
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# CNN Model
cnn_model = Sequential([
    Embedding(input_dim=10000, output_dim=128, input_length=maxlen),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    MaxPooling1D(pool_size=5),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the CNN model
cnn_history = cnn_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val),
                             callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)])

# Evaluate the CNN model
cnn_loss, cnn_accuracy = cnn_model.evaluate(X_test, y_test)
print(f'CNN Test Accuracy: {cnn_accuracy:.2f}')

# LSTM Model
lstm_model = Sequential([
    Embedding(input_dim=10000, output_dim=128, input_length=maxlen),
    LSTM(128, return_sequences=True),
    LSTM(128),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the LSTM model
lstm_history = lstm_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val),
                              callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)])

# Evaluate the LSTM model
lstm_loss, lstm_accuracy = lstm_model.evaluate(X_test, y_test)
print(f'LSTM Test Accuracy: {lstm_accuracy:.2f}')

```

```python
from sklearn.metrics import classification_report

# Predictions for CNN model
cnn_y_pred = cnn_model.predict(X_test)
cnn_y_pred = np.round(cnn_y_pred).flatten()

# Predictions for LSTM model
lstm_y_pred = lstm_model.predict(X_test)
lstm_y_pred = np.round(lstm_y_pred).flatten()

# Classification report for CNN model
print("CNN Model:")
print(classification_report(y_test, cnn_y_pred))

# Classification report for LSTM model
print("LSTM Model:")
print(classification_report(y_test, lstm_y_pred))
```

```python
import matplotlib.pyplot as plt

# Plotting function
def plot_history(history, model_name):
    plt.figure(figsize=(12, 6))

    # Plotting accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plotting loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Plotting CNN model history
plot_history(cnn_history, 'CNN')

# Plotting LSTM model history
plot_history(lstm_history, 'LSTM')
```

```python
# Function to get misclassified samples
def get_misclassified_samples(model, X_data, y_true):
    y_pred = model.predict(X_data)
    y_pred = np.round(y_pred).flatten()
    misclassified_indices = np.where(y_pred != y_true)[0]
    misclassified_samples = [(X_data[i], y_true[i], y_pred[i]) for i in misclassified_indices]
    return misclassified_samples

# Get misclassified samples for CNN model
cnn_misclassified_samples = get_misclassified_samples(cnn_model, X_test, y_test)

# Get misclassified samples for LSTM model
lstm_misclassified_samples = get_misclassified_samples(lstm_model, X_test, y_test)

# Display misclassified samples for CNN model
print("Misclassified samples for CNN model:")
for sample in cnn_misclassified_samples:
    print(f"Text: {tokenizer.sequences_to_texts([sample[0]])[0]}")
    print(f"True Label: {sample[1]}, Predicted Label: {sample[2]}")
    print()

# Display misclassified samples for LSTM model
print("Misclassified samples for LSTM model:")
for sample in lstm_misclassified_samples:
    print(f"Text: {tokenizer.sequences_to_texts([sample[0]])[0]}")
    print(f"True Label: {sample[1]}, Predicted Label: {sample[2]}")
    print()
```
